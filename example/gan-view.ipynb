```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Generator(nn.Module):
    def __init__(self, ngpu):
        super(Generator, self).__init__()
    def forward(self, input):
        return self.main(input)
```

```python
def view_images(network):
    net = torch.load(network)
    with torch.no_grad():
        fixed_noise = torch.randn(9, 100, 1, 1, device='cpu')
        images = net(fixed_noise)

    fig = plt.figure()
    for i in range(9):
        ax = fig.add_subplot(3, 3, i+1)
        ax.imshow(np.transpose(images[i]))
```

```python
# Blobs with high convexity
view_images('gan-mostlyconvex.zip')
```

```python
# "Usual" blobs
view_images('gan-normal.zip')
```

```python
# "Thin" blobs
view_images('gan-thin.zip')
```

```python
# Pretrained on MNIST dataset
view_images('gan-mnist.zip')
```
